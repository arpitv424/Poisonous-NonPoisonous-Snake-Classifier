```
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, Flatten, Dropout, MaxPooling2D
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import numpy as np
import matplotlib. pyplot as plt
from google.colab.patches import cv2_imshow
```

```
train_path="/content/drive/My Drive/SnkCls/Train"
test_path="/content/drive/My Drive/SnkCls/Test"
valid_path="/content/drive/My Drive/SnkCls/valid"
```

```
train=ImageDataGenerator().flow_from_directory(train_path, target_size=(300,300), classes=['N-V','V'], batch_size=7)
```

```
model1 = Sequential([
    Conv2D(96,(3,3), activation='relu', input_shape=(300,300,3)),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(512, activation='relu'),
    Dense(512, activation='relu'),
    Dense(512, activation='relu'),
    Dense(2, activation='softmax')])
```
